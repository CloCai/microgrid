# Problem

UC Berkeley's Center for Effective Global Action (CEGNA), the Center for Information Technology Research in the Interest of Society (CITRIS), and the Electrical Engineering and Computer Science (EECS) department are contributing to the development of rural microgrid projects in rural India and Kenya. A microgrid is a local self-contained electricity generator and those on this project run on solar power and are located in the rural town of Kisii, Kenya.  In many of the villages this is their first and only available power source for their home.  

There is a desire to study the behaviors of these newly connected populations in an effort to help efficiently bring reliable electricity to the remaining 3 billion people on the planet.  This large population suggests there is likely to be a large increase in demand for energy in the near future and energy forcasts have implications for the appropriate scale in energy infrastructure and for the impact on the environment. So, the UC partners are trying to understand the development impacts of rural electrification, as well as its demand and usage behaviors to inform economic experiments with the goal of efficiently expanding electrical usage in a sustainable manner.  

Economists at CEGNA are currently limited to survey data on power usage.  To provide more accurate and timely data, a smart grid approach was taken and every microgrid records power usage metrics and the current amount of credit  for a household.  A prepaid metering system is used where customers use SMS to pay for and monitor their usage on demand. The data is then sent daily over SMS where it is stored and then sent back to Berkeley.  This granularity of data presents opportunities to predict usage behaviors and power usage for particular times of day or week.  

So we approached this opportunity by trying to discover types of users based on their power usage and crediting behavior, whether there is a relationship between power usage and credit transaction behavior, and how peak load times vary among users.


Ensure demand and balance the power load.
